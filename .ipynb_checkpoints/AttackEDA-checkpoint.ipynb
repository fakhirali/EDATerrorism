```python
dates = df["Date"].values
```

```python
days = []
for i in range(len(dates)):
    days.append(dates[i].split("-")[0])
```

```python
dates
```

```python
years = []
for i in range(len(dates)):
    years.append(int(dates[i].split("-")[-1]))
```

```python
mi = min(years)
ma = max(years)
```

```python
counts_per_year = []
for i in range(mi,ma+1):
    counts_per_year.append(years.count(i))
```

```python
plt.plot(range(mi,ma+1),counts_per_year)
plt.plot(2014,years.count(2014),"ro")
plt.plot(2009,years.count(2009),"ro")
plt.plot(2011,years.count(2011),"bo")
```

The first green dot shows when various military operations took place like rah e rast etc. 	General
Ashfaq Parvez Kayani was the army chief at the time.

```python
ndates = []
months = []
for i in range(len(dates)):
    try:
        ndates.append(int(dates[i].split("-")[1].split(" ")[1]))
        months.append(dates[i].split("-")[1].split(" ")[0])
    except:
        ndates.append(int(dates[i].split("-")[2]))
        months.append(dates[i].split("-")[1])
```

```python
all_days = list(set(days))
```

```python
all_days.remove('Thursay')
```

```python
counts_per_days = []
for day in all_days:
    counts_per_days.append(days.count(day))
```

```python
plt.plot(all_days,counts_per_days)
```
