```python
dates = df["Date"].values
```

```python
days = []
for i in range(len(dates)):
    days.append(dates[i].split("-")[0])
```

```python
dates[:5]
```

```python
years = []
for i in range(len(dates)):
    years.append(int(dates[i].split("-")[-1]))
```

```python
min_year = min(years)
max_year = max(years)
```

```python
counts_per_year = []
for i in range(min_year,max_year+1):
    counts_per_year.append(years.count(i))
```

```python
plt.plot(range(min_year,max_year+1),counts_per_year)
plt.plot(2014,years.count(2014),"ro")
plt.plot(2009,years.count(2009),"go")
plt.plot(2011,years.count(2011),"bo")
```

The green dot shows when various military operations took place like rah e rast etc. 	General
Ashfaq Parvez Kayani was the army chief at the time.
The blue dot shows the year when Osama bin Laden was killed.
The red dot shows the year when the peshawar school attack happend.

```python
numdates = []
months = []
for i in range(len(dates)):
    try:
        numdates.append(int(dates[i].split("-")[1].split(" ")[1]))
        months.append(dates[i].split("-")[1].split(" ")[0])
    except:
        numdates.append(int(dates[i].split("-")[2]))
        months.append(dates[i].split("-")[1])
```

```python
gdp = pd.read_csv("Data/GDP.csv")
```

```python
pakgdp_data = gdp.loc[gdp["Country Code"] == "PAK"]
```

```python
pakgdp = pakgdp_data.values[0][4:]
```

```python
yrs = pakgdp_data.columns.values[4:].astype(int)
```

```python
yrs[45:56] #The decade where most of the terrorist attacks occurred 
```

```python
yrs[49] #The year where the highest ammount of attacks occurred
```

```python
plt.figure(figsize=(10,5))
plt.plot(yrs, pakgdp)
plt.plot(yrs[45:56],pakgdp[45:56],"r-")
plt.plot(2009,pakgdp[49],'ro')
```

Even though the economy was consistently rising during that era. In 2009 there was a dip which could be due to increased terrorism.

To further analyze, we could compare the economy with other countries in the region so that we may confirm if the dip is due to terrorism or is just a general dip in the economy due to other external factors.

```python
bangdp = gdp.loc[gdp["Country Name"] == "Bangladesh"].values[0][4:]
plt.figure(figsize=(10,5))
plt.plot(yrs,bangdp)
plt.plot(yrs[45:56],bangdp[45:56],"r-")
plt.plot(2009,bangdp[49],'ro')
plt.title("Bangladesh GDP")
```

Bangladesh does not experience a dip during 2009.

```python
#side by side comparison
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.plot(yrs, pakgdp)
plt.plot(yrs[45:56],pakgdp[45:56],"r-")
plt.plot(2009,pakgdp[49],'ro')
plt.title("Pakistan GDP")

plt.subplot(2,2,2)
plt.plot(yrs,bangdp)
plt.plot(yrs[45:56],bangdp[45:56],"r-")
plt.title("Bangladesh GDP")
plt.plot(2009,bangdp[49],'ro')
```

```python
indgdp = gdp.loc[gdp["Country Code"] == "IND"].values[0][4:]
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.plot(yrs, pakgdp)
plt.plot(yrs[45:56],pakgdp[45:56],"r-")
plt.plot(2009,pakgdp[49],'ro')
plt.title("Pakistan GDP")

plt.subplot(2,2,2)
plt.plot(yrs,indgdp)
plt.plot(yrs[45:56],indgdp[45:56],"r-")
plt.plot(2009,indgdp[49],'ro')
plt.title("India GDP")
```

```python
wldgdp = gdp.loc[gdp["Country Code"] == "WLD"].values[0][4:]
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.plot(yrs, pakgdp)
plt.plot(yrs[45:56],pakgdp[45:56],"r-")
plt.plot(2009,pakgdp[49],'ro')
plt.title("Pakistan GDP")

plt.subplot(2,2,2)
plt.plot(yrs,wldgdp)
plt.plot(yrs[45:56],wldgdp[45:56],"r-")
plt.plot(2009,wldgdp[49],'ro')
plt.title("World GDP")
```

The region does not seejm to have a dip. However looking at the world GDP we can see that there was a massive dip.
